```python
import pandas as pd

csv_data = pd.read_csv('my_data.csv')
copy_df = csv_data

copy_df
```

```python
# 칼럼명 확인하기
print(copy_df.columns)
```

```python

# Unnamed:0 이름 변경 
copy_df.rename(columns={'Unnamed: 0':'number'}, inplace=True)
print(copy_df)

print('\n')

# Unnamed:0(number)열 삭제
copy_df.drop('number', axis=1, inplace=True)
print(copy_df)
```

```python
# name에 해당되는 영문 이름을 한글로 바꾸기
copy_df.replace({'Alice':'앨리스', 'Bob':'밥', 'Charlie':'찰리', 'james':'제임스'}, inplace=True)

copy_df
```

```python
# 천단위 , 찍기
copy_df['salary'] = copy_df['salary'].apply(lambda x: '{:,}'.format(x))

copy_df
```

```python
# 인덱스 제거하기
# copy_df.set_index('name', inplace=True)
copy_df2 = copy_df.style.hide(axis='index')

copy_df2
```

```python
# csv 파일로 저장하기
copy_df.to_csv('my_data_copy.csv', index=False)


df2 = pd.read_csv('my_data_copy.csv')
df2
```
